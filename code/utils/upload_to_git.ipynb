# Upload to git

```python
import time
import os
from glob import glob
import git
```

```python
PATH = "D:\\Users\\Rasilu\\Dokumente\\Schule\\ETH\\Projects\\CIL Project - git\\datasets\\mass_roads\\training\\images"
REPO = "D:\\Users\\Rasilu\\Dokumente\\Schule\\ETH\\Projects\\CIL Project - git"
BLOCK_SIZE = 20
```

```python
def upload_folder(path, repo, origin):
    file_names = glob(path + '/*.png')
    start = time.time()
    for i, f in enumerate(file_names):
        repo.git.add(f)
        if i % BLOCK_SIZE == 0 or i == len(file_names) - 1:
            repo.git.commit('-m', 'upload files' + str(i) + '/' + str(len(file_names)))
            origin.push()

            stop = time.time()
            print("Time for push: " + str(stop - start))
            start = time.time()
```

```python
repo = git.Repo(REPO)
origin = repo.remote(name='origin')
#origin.pull()
```

```python
start_all = time.time()
upload_folder(PATH, repo, origin)
stop_all = time.time()
print("Total time: " + str(stop_all - start_all))
```
